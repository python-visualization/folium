# Plugin: leaflet-search

## Loading some polygon data

```python
import json
with open('data/us-states.json') as data_file:
    states = json.load(data_file)


def getColor(d):
    if d > 1000:
        return '#800026'
    if d > 500:
        return '#BD0026'
    if d > 200:
        return '#E31A1C'
    if d > 100:
        return '#FC4E2A'
    if d > 50:
        return '#FD8D3C'
    if d > 20:
        return '#FEB24C'
    if d > 10:
        return '#FED976'
    else:
        return '#FFEDA0'


for state in states['features']:
    state['style'] = {'color': getColor(state['properties']['density'])}
```

## Polygon search layer

```python
import os
import folium
from folium import plugins

m = folium.Map(
    location=[37.8, -96],
    tiles='cartodbpositron',
    zoom_start=5)

plugins.Search(states, search_zoom=6, geom_type='Polygon').add_to(m)

m.save(os.path.join('results', 'Plugins_search_polygon.html'))

m
```

## Loading some Points data

```python
import json

bars_path = os.path.join('data', 'bars.json')

bars = json.load(open(bars_path))
```

## Points search layer

```python
m = folium.Map(
    location=[41.575330, 13.102411],
    tiles='cartodbpositron',
    zoom_start=9)

plugins.Search(bars, search_zoom=20).add_to(m)

m.save(os.path.join('results', 'Plugins_search_points.html'))

m
```
