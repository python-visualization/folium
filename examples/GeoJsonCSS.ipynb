```python
import os
import folium

print(folium.__version__)
```

### Default, no css

```python
m = folium.Map(
    location=[55.75258, 37.61489],
    zoom_start=14,
    tiles = 'cartodbpositron'
)

fname = os.path.join('data', 'css_demo.geojson')

folium.GeoJson(fname).add_to(m)

m
```

### Using the css in the metadata

```python
m = folium.Map(
    location=[55.75258, 37.61489],
    zoom_start=14,
    tiles = 'cartodbpositron'
)


folium.GeoJsonCss(fname).add_to(m)

m
```
